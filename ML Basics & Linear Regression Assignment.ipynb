**Part A – Theoretical Questions (Machine Learning & Linear Regression)**

**Q1. What is Machine Learning? Explain the main types of Machine Learning with examples.**

ANS. Machine Learning (ML) is a subset of Artificial Intelligence (AI) where computers learn from data and improve their performance over time without being explicitly programmed. Instead of following hard-coded rules, ML algorithms identify patterns in large datasets to make decisions or predictions.
- The main types of Machine Learning:
- Supervised Learning:The model is trained on labeled data, meaning the input data is paired with the correct output. The model learns the relationship between inputs and outputs to predict outcomes for unseen data.
Example: Email Spam Filtering. The system is trained on thousands of emails labeled as "spam" or "not spam." Once trained, it can automatically detect and route new spam emails to the junk folder.
- Unsupervised Learning:The model is provided with unlabeled data without any predefined answers. The algorithm is tasked with discovering hidden patterns, structures, or groupings within the data on its own.
Example: Customer Segmentation. An e-commerce platform groups buyers based on purchasing habits and browsing behavior without prior knowledge of what those groups are, helping target specific demographics.
-  Reinforcement Learning:The model learns to make decisions through trial and error. It interacts with an environment and receives a reward for taking positive actions or a penalty for mistakes, aiming to maximize its cumulative reward over time.
Example: Autonomous Vehicles. A self-driving car learns to navigate traffic and park by being rewarded for safe driving and penalized for collisions or traffic violations.
- Semi-Supervised Learning:A hybrid approach that uses a small amount of labeled data and a much larger set of unlabeled data. It is highly useful when acquiring labeled data is expensive or time-consuming, allowing the model to learn the basic structure from the unlabeled data and refine it with the labeled data.
Example: Image Recognition. Labeling every single photo in a massive dataset (e.g., millions of X-rays) is tedious, so a few are labeled by experts, and the algorithm learns the rest by clustering similar unlabeled images.

**Q2. What is Supervised Learning? Explain the difference between Regression and Classification.**

ANS. Supervised learning is a machine learning technique where a model learns to map inputs to specific outputs based on labeled training data. The algorithm analyzes known input-output pairs to infer patterns and make predictions on unseen data.
- The difference between Regression and Classification:
**Regression**
- Output Type: Continuous (numerical values).
- Goal: To predict a continuous, real-valued number based on the relationships found in the data.
- Common Examples: Predicting housing prices, estimating future sales revenue, or forecasting tomorrow's exact temperature.
- Helpful Algorithms: Linear Regression, Ridge Regression, and Random Forest Regressors.
**Classification**
- Output Type: Categorical (discrete labels).
- Goal: To predict which distinct class or category an input belongs to.
- Common Examples: Email spam detection (spam or not spam), image recognition (cat, dog, or bird), and disease diagnosis (positive or negative).
- Helpful Algorithms: Logistic Regression, Decision Trees, and Support Vector Machines.

**Q3. What is Linear Regression? Write the Linear Regression equation and explain intercept and coefficient.**

ANS.Linear Regression is a supervised machine learning algorithm used for predictive analysis. It models the relationship between a dependent target variable and one or more independent predictor variables by fitting a straight line to the observed data.
**The Linear Regression Equation**
For a simple linear regression (one independent variable), the equation is represented as:
- (Y = a + bX)
**Where:**
- (Y) = The dependent variable (the outcome you are trying to predict).
- (X) = The independent variable (the predictor variable).
- (b) = The coefficient or slope.
- (a) = The intercept.
**Explaining Intercept and Coefficient**
- The Intercept: This is the value of the dependent variable (Y) when the independent variable (X) is exactly 0. Geometrically, it is the point where the regression line crosses the y-axis. For example, if (X) is study hours and (Y) is test score, the intercept represents the expected score of a student who studied for zero hours.
- The Coefficient / Slope: This represents the rate of change in the dependent variable (Y) for every one-unit increase in the independent variable (X). It determines the steepness and direction of the regression line. For instance, a coefficient of 5 means that for every additional hour spent studying (X), the test score (Y) increases by 5 points.

**Q4. Explain the following evaluation metrics used in Linear Regression:**
- SSE
- MSE
- RMSE
- R² Score
- Adjusted R²

ANS. Evaluation metrics in linear regression quantify how well your model's predictions match actual data. They determine error magnitude and overall model fit.
**Here is a breakdown of the key evaluation metrics:**
**SSE (Sum of Squared Errors)**
- SSE measures the total deviation of the predicted values from the actual observed values.
- Interpretation: A lower SSE indicates a better fit. However, because it is the sum of squared differences, its value increases as the dataset gets larger, making it difficult to use as a standalone metric for comparing datasets of different sizes.
**MSE (Mean Squared Error)**
- MSE is the average of the squared errors between the predicted and actual values.
- Interpretation: Because of the squaring effect, MSE heavily penalizes large errors (outliers). Lower values denote better accuracy, though it is reported in squared units of the target variable.
**RMSE (Root Mean Squared Error)**
- RMSE is the square root of the MSE. It represents the standard deviation of the residuals (prediction errors).
- Interpretation: RMSE is one of the most widely used metrics. By taking the root of the MSE, it returns the error to the original scale of your target variable, making it highly interpretable. Like MSE, a lower RMSE score indicates a more accurate model.
**R^2 Score (Coefficient of Determination)**
- R^2 represents the proportion of the variance in the dependent variable that is predictable and explained by the independent variables.
- Interpretation: It typically ranges from 0 to 1. An R^2 score of 0.80 means that 80% of the variation in the target is explained by the model. Higher is generally better, but R^2 has a drawback: it will always increase or stay the same when new independent variables are added, even if those variables are completely irrelevant.

**Q5. Explain the Machine Learning workflow steps from data collection to model evaluation.**

ANS. The Machine Learning (ML) workflow is an iterative, structured sequence of steps used to transform raw data into a predictive model. It begins with understanding the problem and progresses through data handling and training to final performance assessment.
**Data Collection**
This is the foundational step where relevant raw data is gathered from various sources such as databases, APIs, web scraping, or IoT sensors. The quality and quantity of the data collected at this stage directly determine the potential effectiveness of the final model.
**Data Preprocessing (Preparation)**
Raw data is rarely ready for training. This step involves cleaning and formatting to ensure the data is reliable:
- Data Cleaning: Identifying and handling missing values, removing duplicates, and correcting inconsistent formats.
- Data Transformation: Normalizing or scaling numerical values to ensure different features are on a similar scale.
- Exploratory Data Analysis (EDA): Visualizing and analyzing the data to identify underlying patterns and distributions.
**Feature Engineering**
Feature engineering is the process of selecting, modifying, or creating new variables (features) that help the model learn more effectively. This might include:
- Feature Selection: Dropping irrelevant or redundant columns to reduce complexity.
- Feature Extraction: Creating new insights from existing data, such as extracting "hour of the day" from a timestamp.
**Data Splitting**
Before training, the dataset is typically divided into three subsets to prevent "memorization" (overfitting):
- Training Set: Used to teach the model.
- Validation Set: Used to fine-tune hyperparameters and select the best version of the model.
- Test Set: A "blind" set used only at the very end to provide an unbiased evaluation of the final model.
**Model Selection & Training**
- Model Selection: Choosing an appropriate algorithm (e.g., Linear Regression, Decision Trees, or Neural Networks) based on the problem type, such as classification or regression.
- Model Training: The core process where the selected algorithm "learns" by finding patterns and relationships within the training data.
**Model Evaluation**
The final step in this sequence assesses how well the model generalizes to new, unseen data using the test set. Common metrics include:
- Classification Metrics: Accuracy, Precision, Recall, F1-Score, and Confusion Matrix.
- Regression Metrics: Mean Absolute Error (MAE) and Mean Squared Error (MSE).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics 
```

**Q6. Load the Boston Housing dataset and perform basic data exploration using:

info()

```python
df = pd.read_csv('BostonHousing.csv')
df.head()
df.info()
df.describe()
```

**Q7. Perform Exploratory Data Analysis:
Create pairplot
Create correlation heatmap
Identify which features are highly correlated with house price (medv)**

```python
# Pairplot
sns.pairplot(df[['rm', 'lstat', 'ptratio', 'medv']])
plt.show()
```

```python
# Correlation matrix
corr_matrix = df.corr()

# Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Correlation with target variable
print(corr_matrix['medv'].sort_values(ascending=False))
```

**Interpretation**
- rm (0.695) has a strong positive correlation with house prices. Houses with more rooms tend to have higher prices.
- lstat (-0.738) has the strongest negative correlation with house prices. As the percentage of lower-status population increases, house prices tend to decrease.
- ptratio (-0.508) shows a moderate negative correlation. Areas with higher pupil-teacher ratios generally have lower house prices.
- tax (-0.469) and indus (-0.484) also have moderate negative correlations with house prices.
- zn (0.360) and b (0.333) have moderate positive correlations with house prices.

**Q8. Select independent variables (features) and dependent variable (target).
Target variable: medv (house price)**

```python
# Independent variables
X = df.drop('medv', axis=1)
print("Independent Variables (X):")
print(X.head())

# Dependent variable
y = df['medv']
print("Dependent Variable (y):")
print(y.head())
```

**Q9. Split the dataset into training and testing sets (80% training, 20% testing).**

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)
# Shape of datasets
print(X_train.shape)
print(X_test.shape)
```

**Q10. Train a Linear Regression model and perform the following:
Make predictions
Calculate MAE
Calculate MSE
Calculate RMSE
Calculate R² Score
Calculate Adjusted R²
Write the Linear Regression equation
Interpret the model coefficients**

```python
# Create model
model = LinearRegression()

# Train model
model.fit(X_train, y_train)
```

```python
# Predictions
y_pred = model.predict(X_test)

print(y_pred[:5])
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)

print("MAE:", mae)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)
```

```python
rmse = np.sqrt(mse)

print("RMSE:", rmse)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

print("R² Score:", r2)
```

```python
n = X_test.shape[0]   
p = X_test.shape[1]  

adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))

print("Adjusted R²:", adjusted_r2)
```

```python
# Intercept
print("Intercept:", model.intercept_)

# Coefficients
coeff_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})

print(coeff_df)
```

**Coefficient Interpretation:**
- rm has a positive coefficient, meaning houses with more rooms tend to have higher prices.
- lstat has a negative coefficient, meaning areas with a higher percentage of lower-status population tend to have lower house prices.
- crim has a negative coefficient, indicating that higher crime rates are associated with lower house prices.
- zn has a positive coefficient, suggesting that houses in large residential zones tend to be more expensive.
